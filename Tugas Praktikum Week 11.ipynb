```python
# courses collection (dipakai di semua homework)
from pymongo import MongoClient, InsertOne

client = MongoClient("mongodb://localhost:27017/")
db = client['university_db']
courses_collection = db['courses']

# Hapus data lama (opsional)
courses_collection.delete_many({})

# Dataset contoh courses
operations = [
    InsertOne({'course': 'Math 101', 'enrollments': 30, 'department': 'Mathematics'}),
    InsertOne({'course': 'CS 102', 'enrollments': 28, 'department': 'Computer Science'}),
    InsertOne({'course': 'CS 103', 'enrollments': 22, 'department': 'Computer Science'}),
    InsertOne({'course': 'History 201', 'enrollments': 20, 'department': 'History'}),
    InsertOne({'course': 'Physics 202', 'enrollments': 15, 'department': 'Physics'}),
    InsertOne({'course': 'Fisika Dasar', 'enrollments': 25, 'department': 'Fisika'}),
    InsertOne({'course': 'Teknik Fisika 101', 'enrollments': 18, 'department': 'Teknik Fisika'})
]

courses_collection.bulk_write(operations)
print("Dataset courses inserted successfully.")
```

```python
# students collection (dipakai di Homework 3)
students_collection = db['students']
students_collection.delete_many({})

# Dataset contoh students
operations = [
    InsertOne({'name': 'Senandung', 'course': 'Math 101'}),
    InsertOne({'name': 'Wisnu', 'course': 'CS 102'}),
    InsertOne({'name': 'Lis', 'course': 'CS 102'}),
    InsertOne({'name': 'Kiandra', 'course': 'CS 103'}),
    InsertOne({'name': 'Rezl', 'course': 'Fisika Dasar'}),
    InsertOne({'name': 'Yoko', 'course': 'Teknik Fisika 101'})
]

students_collection.bulk_write(operations)
print("Dataset students inserted successfully.")
```

```python
# Homework 1: Aggregation to get count of courses per department
pipeline = [{'$group': {'_id': '$department', 'course_count': {'$sum': 1}}}]

# Jalankan pipeline agregasi dan tampilkan hasil
print("Homework 1 – Count of courses per department:")
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Homework 2: Filter courses di 'Computer Science' dengan enrollments > 25, lalu hitung jumlahnya per department
pipeline = [
    {'$match': {'department': 'Computer Science', 'enrollments': {'$gt': 25}}},
    {'$group': {'_id': '$department', 'course_count': {'$sum': 1}}}
]
# Jalankan pipeline agregasi dan tampilkan hasil
print("\nHomework 2 – Courses > 25 enrollments in Computer Science:")
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Homework 3: Join courses dengan students collection berdasarkan field 'course' menggunakan $lookup
pipeline = [
    {'$lookup': {'from': 'students', 'localField': 'course', 'foreignField': 'course', 'as': 'enrolled_students'}}
]

# Jalankan pipeline agregasi dan tampilkan hasil
print("\nHomework 3 – Courses with enrolled students:")
for result in courses_collection.aggregate(pipeline):
    print(result)
```
