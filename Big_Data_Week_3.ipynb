```python
# Tugas 1
print("=== Tugas 1: Data Awal ===")
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('HandsOnPertemuan3').getOrCreate()

data = [('James', 'Sales', 3000),
        ('Michael', 'Sales', 4600),
        ('Robert', 'Sales', 4100),
        ('Maria', 'Finance', 3000),
        ('Dudung', 'Analyst', 15000),]

columns = ['EmployeeName', 'Department', 'Salary']

df = spark.createDataFrame(data, schema=columns)
df.show()

# Tugas 2
print("=== Tugas 2.1: Select EmployeeName dan Salary ===")
df.select('EmployeeName', 'Salary').show()

print("=== Tugas 2.2: Filter Salary > 3000 ===")
df.filter(df['Salary'] > 3000).show()

print("=== Tugas 2.3: GroupBy Department lalu rata-rata Salary ===")
df.groupBy('Department').avg('Salary').show()

# Tugas 3
from pyspark.sql.functions import col
print("=== Tugas 3: Tambah kolom SalaryBonus dan TotalCompensation ===")
df = df.withColumn('SalaryBonus', col('Salary') * 0.1)
df = df.withColumn('TotalCompensation', col('Salary') + col('SalaryBonus'))
df.show()

# Tugas 4
from pyspark.sql.window import Window
from pyspark.sql import functions as F
print("=== Tugas 4: Rank berdasarkan Salary tiap Department ===")
windowSpec = Window.partitionBy('Department').orderBy('Salary')
df.withColumn('Rank', F.rank().over(windowSpec)).show()
```

```python
# --- 1️⃣ Upload file CSV dari Kaggle atau lokal ---
from google.colab import files
uploaded = files.upload()  # pilih crocodile_dataset.csv

# --- 2️⃣ Setup PySpark ---
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, avg

spark = SparkSession.builder.appName("Analisis Dataset Buaya").getOrCreate()

# --- 3️⃣ Load dataset ke PySpark ---
csv_file = list(uploaded.keys())[0]   # ambil nama file yang diupload
df = spark.read.csv(csv_file, header=True, inferSchema=True)

# --- 4️⃣ Eksplorasi awal dataset ---
print("Skema dataset:")
df.printSchema()

print("\nContoh 5 baris data:")
df.show(5)

print("\nJumlah baris:", df.count())
print("Jumlah kolom:", len(df.columns))
print("Nama kolom:", df.columns)

# --- 5️⃣ Manipulasi Data ---
# Misalnya: hitung jumlah data per spesies buaya (kalau ada kolom 'species')
if "species" in df.columns:
    species_count = df.groupBy("species").agg(count("*").alias("jumlah"))
    print("\nJumlah data per spesies:")
    species_count.show()

# Misalnya: hitung rata-rata panjang buaya (kalau ada kolom 'length')
if "length" in df.columns:
    avg_length = df.agg(avg(col("length")).alias("Rata-rata panjang"))
    print("\nRata-rata panjang buaya:")
    avg_length.show()

# --- 6️⃣ Simpan hasil ke file baru ---
output_path = "/content/crocodile_processed.csv"
df.write.csv(output_path, header=True, mode="overwrite")
print("\nDataset berhasil diproses dan disimpan di:", output_path)

# --- 7️⃣ Kesimpulan & Eksplorasi Lanjut ---
# - Dataset berhasil dimuat menggunakan PySpark
# - Data dapat dimanipulasi (groupBy, aggregasi, filtering, dsb.)
# - Hasil bisa disimpan kembali ke CSV
# Eksplorasi lanjut: bisa menambahkan visualisasi, analisis korelasi, atau model ML
```
